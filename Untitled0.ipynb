```python
# Abhishek Dubey
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
data = {
    'Vehicle Class': [
        'FOUR WHEELER (INVALID CARRIAGE)', 'HEAVY GOODS VEHICLE', 'HEAVY MOTOR VEHICLE',
        'HEAVY PASSENGER VEHICLE', 'LIGHT GOODS VEHICLE', 'LIGHT MOTOR VEHICLE',
        'LIGHT PASSENGER VEHICLE', 'MEDIUM GOODS VEHICLE', 'MEDIUM MOTOR VEHICLE',
        'MEDIUM PASSENGER VEHICLE', 'OTHER THAN MENTIONED ABOVE', 'THREE WHEELER(NT)',
        'THREE WHEELER(T)', 'TWO WHEELER (INVALID CARRIAGE)', 'TWO WHEELER(NT)', 'TWO WHEELER(T)'
    ],
    'Total Registration': [
        "21,346", "58,70,865", "1,02,965", "8,28,189", "1,02,49,591",
        "6,50,61,773", "43,43,410", "8,75,789", "1,94,600", "3,25,015",
        "11,26,398", "6,79,804", "1,07,08,473", "1,10,788", "27,49,71,646", "1,29,181"
    ]
}

df = pd.DataFrame(data)

# Clean the Total Registration column - remove commas and convert to numeric
df['Total Registration'] = df['Total Registration'].str.replace(',', '').astype(float)

# Set up the plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
# Create categories
def categorize_vehicle(vehicle_class):
    vehicle_class_lower = vehicle_class.lower()
    if 'two wheeler' in vehicle_class_lower:
        return 'Two Wheelers'
    elif 'three wheeler' in vehicle_class_lower:
        return 'Three Wheelers'
    elif 'light' in vehicle_class_lower:
        return 'Light Vehicles'
    elif 'medium' in vehicle_class_lower:
        return 'Medium Vehicles'
    elif 'heavy' in vehicle_class_lower:
        return 'Heavy Vehicles'
    elif 'four wheeler' in vehicle_class_lower:
        return 'Four Wheelers'
    else:
        return 'Others'
df['Vehicle Category'] = df['Vehicle Class'].apply(categorize_vehicle)
category_totals = df.groupby('Vehicle Category')['Total Registration'].sum().sort_values(ascending=False)

# Category-wise Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(category_totals.index, category_totals.values)
plt.title('Total Registration by Vehicle Category', fontsize=16, fontweight='bold')
plt.xlabel('Vehicle Category', fontsize=12)
plt.ylabel('Total Registration', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:,.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Abhishek Dubey:
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Load the Dataset
file_path = "Indian automoble buying behavour study 1.0.csv"
df = pd.read_csv(file_path)

# Quick preview
df.head()
# Convert numeric columns
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Total Salary'] = pd.to_numeric(df['Total Salary'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Drop rows with missing or invalid data
df_clean = df.dropna(subset=['Total Salary', 'Price'])

# Create income group
df_clean['Income Group'] = pd.cut(
    df_clean['Total Salary'],
    bins=[0, 500000, 1000000, 1500000, 2000000, 3000000],
    labels=['<5L', '5L-10L', '10L-15L', '15L-20L', '>20L']
)
df_clean.head()
# Data Scaling for Clustering
X = df_clean[['Total Salary', 'Price']].values

# Standardize data (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Optimal Number of Clusters (Elbow Method)
inertia = []
k_values = range(1, 8)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(7,4))
plt.plot(k_values, inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

#  Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_clean['Cluster'] = kmeans.fit_predict(X_scaled)

# Cluster centers (unscaled)
centers = scaler.inverse_transform(kmeans.cluster_centers_)

# Add cluster labels
df_clean.head()

# Visualize Market Segments
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_clean, x='Total Salary', y='Price', hue='Cluster', palette='Set2', s=80)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("Market Segmentation: Income vs Vehicle Price")
plt.xlabel("Total Family Annual Income (INR)")
plt.ylabel("Vehicle Price (INR)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Segment Interpretation
segment_summary = df_clean.groupby('Cluster')[['Total Salary', 'Price']].mean()
segment_summary['Count'] = df_clean['Cluster'].value_counts().sort_index()
segment_summary
for i, row in segment_summary.iterrows():
    print(f"Segment {i}:")
    print(f" - Average Income: {row['Total Salary']:.0f}")
    print(f" - Average Vehicle Price: {row['Price']:.0f}")
    print(f" - Customers in Segment: {int(row['Count'])}")
    print(" - Possible Insight:", end=" ")

    if row['Total Salary'] < 1000000:
        print("Budget-conscious buyers; prefer small cars or entry-level EVs.")
    elif 1000000 <= row['Total Salary'] < 2000000:
        print("Mid-range buyers; interested in compact SUVs or mid-segment EVs.")
    else:
        print("High-income premium buyers; open to performance or luxury EVs.")
    print()
```

```python
# Bandaru Harika Sai Sri and Yash Salvi
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# ========== OUTPUT DIRECTORIES ==========
output_dir = "EV_Output"
fig_dir = os.path.join(output_dir, "figures")
os.makedirs(output_dir, exist_ok=True)
os.makedirs(fig_dir, exist_ok=True)

# ========== INPUT DATA ==========
pcs_data = {
    "Maharashtra": 3079, "Delhi": 1886, "Karnataka": 1041, "Kerala": 852, "Tamil Nadu": 643,
    "Uttar Pradesh": 582, "Rajasthan": 500, "Telangana": 481, "Gujarat": 476, "Haryana": 377,
    "Madhya Pradesh": 341, "Andhra Pradesh": 327, "West Bengal": 318, "Odisha": 198, "Punjab": 158,
    "Chhattisgarh": 149, "Jharkhand": 135, "Bihar": 124, "Goa": 113, "Assam": 86, "Uttarakhand": 76,
    "Jammu & Kashmir": 47, "Himachal Pradesh": 44, "Pondicherry": 23, "Meghalaya": 21, "Tripura": 18,
    "Manipur": 17, "Chandigarh": 12, "Arunachal Pradesh": 9, "Nagaland": 6, "Andaman & Nicobar": 3,
    "Sikkim": 2, "D&D and DNH": 1, "Lakshadweep": 1
}

population_proxies = {
    "Maharashtra": 123.0, "Delhi": 31.0, "Karnataka": 68.0, "Kerala": 35.0, "Tamil Nadu": 77.0,
    "Uttar Pradesh": 237.0, "Rajasthan": 81.0, "Telangana": 39.0, "Gujarat": 69.0, "Haryana": 29.0,
    "Madhya Pradesh": 85.0, "Andhra Pradesh": 53.0, "West Bengal": 101.0, "Odisha": 46.0, "Punjab": 30.0,
    "Chhattisgarh": 30.0, "Jharkhand": 38.0, "Bihar": 124.0, "Goa": 1.5, "Assam": 36.0,
    "Uttarakhand": 11.0, "Jammu & Kashmir": 13.5, "Himachal Pradesh": 7.4, "Pondicherry": 1.4,
    "Meghalaya": 3.2, "Tripura": 4.0, "Manipur": 3.0, "Chandigarh": 1.2, "Arunachal Pradesh": 1.5,
    "Nagaland": 2.2, "Andaman & Nicobar": 0.4, "Sikkim": 0.6, "D&D and DNH": 0.6, "Lakshadweep": 0.07
}

# ========== DATAFRAME CREATION ==========
df = pd.DataFrame({
    "State": list(pcs_data.keys()),
    "PCS_Count": list(pcs_data.values())
})

df["Population_Million"] = df["State"].map(population_proxies)
df["PCS_per_Million"] = df.apply(
    lambda r: r["PCS_Count"] / r["Population_Million"] if pd.notna(r["Population_Million"]) and r["Population_Million"] > 0 else np.nan,
    axis=1
)

# ========== BASIC STATISTICS ==========
stats = {
    "Total_PCS": int(df["PCS_Count"].sum()),
    "Mean_PCS": round(df["PCS_Count"].mean(), 2),
    "Median_PCS": round(df["PCS_Count"].median(), 2),
    "States_Analyzed": df.shape[0]
}

# ========== CLUSTERING ==========
try:
    from sklearn.cluster import KMeans
    X = df[["PCS_Count", "PCS_per_Million"]].fillna(0)
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    df["Cluster"] = kmeans.fit_predict(X)
    cluster_type = "KMeans (3 clusters)"
except Exception:
    df["Cluster"] = pd.qcut(df["PCS_Count"], q=3, labels=[0, 1, 2])
    cluster_type = "Quantile-based (3 groups)"

# ========== GRAPH 1: TOP 10 STATES ==========
plt.figure(figsize=(10, 6))
top10 = df.sort_values("PCS_Count", ascending=False).head(10)
plt.bar(top10["State"], top10["PCS_Count"], color="steelblue")
plt.title("Top 10 States by PCS Count", fontsize=14, weight='bold')
plt.xlabel("State", fontsize=12)
plt.ylabel("Number of PCS", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig(os.path.join(fig_dir, "Top10_PCS.png"))
plt.show()  # Display

# ========== GRAPH 2: SCATTER PLOT ==========
plt.figure(figsize=(8, 6))
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]
for c in sorted(df["Cluster"].unique()):
    sub = df[df["Cluster"] == c]
    plt.scatter(sub["PCS_Count"], sub["PCS_per_Million"], label=f"Cluster {c}", s=80, alpha=0.8, color=colors[int(c) % len(colors)])
plt.title("PCS Count vs PCS per Million Population", fontsize=14, weight='bold')
plt.xlabel("PCS Count", fontsize=12)
plt.ylabel("PCS per Million (approx)", fontsize=12)
plt.legend(title="Cluster", fontsize=10)
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig(os.path.join(fig_dir, "Scatter_PCS_Cluster.png"))
plt.show()  # Display

# ========== GRAPH 3: BAR CHART OF PCS PER MILLION ==========
plt.figure(figsize=(12, 6))
sorted_df = df.sort_values("PCS_per_Million", ascending=False).head(15)
plt.bar(sorted_df["State"], sorted_df["PCS_per_Million"], color="mediumseagreen")
plt.title("Top 15 States by PCS per Million Population", fontsize=14, weight='bold')
plt.xlabel("State", fontsize=12)
plt.ylabel("PCS per Million", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig(os.path.join(fig_dir, "Top15_PCS_per_Million.png"))
plt.show()  # Display

# ========== EXPORT DATA ==========
csv_path = os.path.join(output_dir, "EV_PCS_Analysis.csv")
df.to_csv(csv_path, index=False)

# ========== CREATE PDF REPORT ==========
pdf_path = os.path.join(output_dir, "EV_PCS_Report.pdf")
with PdfPages(pdf_path) as pdf:
    # Cover Page
    plt.figure(figsize=(8.27, 11.69))
    plt.axis("off")
    plt.text(0.5, 0.85, "Electric Vehicle Infrastructure Report", ha="center", fontsize=20, weight="bold")
    plt.text(0.5, 0.78, "Market Segmentation Analysis (Public Charging Stations)", ha="center", fontsize=11)
    plt.text(0.5, 0.7, f"Total PCS: {stats['Total_PCS']:,}", ha="center", fontsize=12)
    plt.text(0.5, 0.66, f"Mean PCS per State: {stats['Mean_PCS']}", ha="center", fontsize=12)
    plt.text(0.5, 0.62, f"Clustering Method: {cluster_type}", ha="center", fontsize=10)
    plt.text(0.5, 0.58, f"Total States Analyzed: {stats['States_Analyzed']}", ha="center", fontsize=10)
    plt.text(0.5, 0.48, "Generated by Internship EV Data Analysis Project", ha="center", fontsize=10)
    pdf.savefig()
    plt.close()

    # Add saved plots
    for img_name in ["Top10_PCS.png", "Scatter_PCS_Cluster.png", "Top15_PCS_per_Million.png"]:
        img = plt.imread(os.path.join(fig_dir, img_name))
        plt.figure(figsize=(8.27, 11.69))
        plt.imshow(img)
        plt.axis("off")
        pdf.savefig()
        plt.close()

    # Add sample data table
    plt.figure(figsize=(8.27, 11.69))
    plt.axis("off")
    table = plt.table(cellText=df.head(25).values, colLabels=df.columns, loc="center", cellLoc="center")
    table.auto_set_font_size(False)
    table.set_fontsize(7)
    table.scale(1.2, 1.2)
    pdf.savefig()
    plt.close()

print("\n Report Generated Successfully")
print(f" CSV: {csv_path}")
print(f" PDF: {pdf_path}")
print(f" Figures saved in: {fig_dir}")
print("\nPreview of DataFrame:")
print(df.head(10))
```
